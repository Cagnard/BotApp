```python
#str_date end_date or_city dst_city budget DateArr DateDep VilleDep VilleArr Budget
def transName(name):
    if name == 'str_date':
        return 'DateArr'
    elif name == 'end_date':
        return 'DateDep'
    elif name == 'or_city':
        return 'VilleDep'
    elif name == 'dst_city':
        return 'VilleArr'
    elif name == 'budget':
        return 'Budget'
    else:
        return None
```

```python
def get_example_label(utterance, entity_name, value):
    utterance = utterance.lower()
    value = value.lower()
    if utterance.find(value) >=0:
        return {
            'entity_name': entity_name,
            'start_char_index': utterance.find(value),
            'end_char_index': utterance.find(value) + len(value)
        }
    else:
        return None
```

```python
def convert_entity_batch(data):
    ent = {
        'entity': data['entity_name'],
        'startPos': data['start_char_index'],
        'endPos': data['end_char_index']
    }
    return ent
```

```python
def convert_batch(data):
    utt = {
        'text': data['text'],
        'intent': data['intent_name'],
        'entities': [convert_entity_batch(ent) for ent in data['entity_labels']]
    }
    return utt
```
